```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
#data
x=np.array([-1.0,0.0,1.0,2.0,3.0,4.0],dtype=float)
y=np.array([-3.0,-1.0,1.0,3.0,5.0,7.0],dtype=float)
```

```python
#model

model=Sequential([Dense(units=1,input_shape=[1])])
model.compile(optimizer='sgd',loss='mean_squared_error')
model.fit(x,y,epochs=500)
plt.scatter(x,y)
plt.plot(x,model.predict(x))
plt.show()
```

```python
#giving value of Y when X=10 can change according to own needs
print(model.predict([10.0]))
```

***Making changes in the model***

```python
model=Sequential([Dense(units=4,input_shape=[1])])
model.compile(optimizer='sgd',loss='mean_squared_error')
model.fit(x,y,epochs=500)
plt.scatter(x,y)
plt.plot(x,model.predict(x))
plt.show
```

```python
#giving value of Y when X=10 can change according to own needs
print(model.predict([10.0]))
```

```python
#model

model=Sequential([Dense(units=1,input_shape=[1]),
                  layers.Dense(4, activation='softmax')])
model.compile(optimizer='sgd',loss='mean_squared_error')
model.fit(x,y,epochs=500)
plt.scatter(x,y)
plt.plot(x,model.predict(x))
plt.show()
#giving value of Y when X=10 can change according to own needs
print(model.predict([10.0]))
```

# **Improving the model**

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(x,y, validation_data=(x, y), epochs=500, callbacks=[early_stopping])
model.predict([10.0])
```

```python
from tensorflow.keras.layers import Dropout
model = Sequential([
    Dense(128, activation='relu', input_shape=([1])),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x, y, epochs=100, validation_data=(x, y))
```
